**Part 1.3 b - Share insights on correlation**

Let us classify correlation relationship between features as follows:

1.	Strong Positive Correlation: A strong upward (positive) linear relationship between two variables. This typically includes correlation values close to 1.
•	Correlation value >0.7>0.7
2.	Weak Positive Correlation: A weaker upward (positive) linear relationship between two variables.
•	Correlation value ≤0.7≤0.7 and >0.3>0.3
3.	Negligible Correlation: Very weak or no linear relationship between two variables.
•	Correlation value ≤0.3≤0.3 and ≥−0.3≥−0.3
4.	Weak Negative Correlation: A weaker downward (negative) linear relationship between two variables.
•	Correlation value <−0.3<−0.3 and ≥−0.7≥−0.7
5.	Strong Negative Correlation: A strong downward (negative) linear relationship between two variables. This typically includes correlation values close to -1.
•	Correlation value <−0.7<−0.7



```python
# The function compute_and_classify_correlation processes each pair of features and prints out the correlation value and its corresponding category.
from scipy.stats import pearsonr

def compute_and_classify_correlation_strength(df):
    # List of features
    features = df.columns[:-1]  # Excluding 'Class' column
    results = []

    # Compute the correlation for each pair of features
    for i in range(len(features)):
        for j in range(i+1, len(features)):
            corr_coeff, _ = pearsonr(df[features[i]], df[features[j]])
            results.append((features[i], features[j], corr_coeff))

    # Sort results based on absolute correlation values
    sorted_results = sorted(results, key=lambda x: abs(x[2]), reverse=True)

    # Classify and print
    for pair in sorted_results:
        feature1, feature2, corr_value = pair
        if corr_value > 0.7:
            corr_type = "Strong Positive"
        elif 0.3 < corr_value <= 0.7:
            corr_type = "Weak Positive"
        elif -0.3 < corr_value <= 0.3:
            corr_type = "Negligible"
        elif -0.7 < corr_value <= -0.3:
            corr_type = "Weak Negative"
        else:
            corr_type = "Strong Negative"

        print(f"Features: {feature1} & {feature2} - Correlation Value: {corr_value:.2f} - Type: {corr_type}")

# Calling the function
compute_and_classify_correlation_strength(df_combined)

```

**Part 1.4 c - Train a Supervised Learning Classification base model using KNN classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initializing the KNN classifier with default settings (k=5)
knn_classifier = KNeighborsClassifier()

# Training the KNN classifier
knn_classifier.fit(X_train, Y_train)

# Predicting the classes for the test set
Y_pred = knn_classifier.predict(X_test)

# Evaluating the performance of the KNN classifier
accuracy = accuracy_score(Y_test, Y_pred)
class_report = classification_report(Y_test, Y_pred)


print("accuracy:",accuracy)
print("class_report:\n",class_report)
```

**Part 1.5 a - Experiment with various parameters to improve performance of the base model**

```python
from sklearn.model_selection import GridSearchCV

# Parameters to be tuned
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski'],
    'p': [1, 2]  # p=1 for manhattan, p=2 for euclidean
}

# Using GridSearchCV to find the best parameters
knn_grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
knn_grid.fit(X_train, Y_train)

# Best parameters found
best_params = knn_grid.best_params_
best_accuracy = knn_grid.best_score_

best_params, best_accuracy
```

After experimenting with various parameters using GridSearchCV for the KNN classifier, the best parameters found are:

Metric: Euclidean

Number of Neighbors (k): 11

p: 1 (which aligns with the Manhattan distance when metric is 'manhattan', but here it's redundant as the best metric is 'euclidean')

Weights: Distance (which means closer neighbors will have a higher influence)

The best accuracy achieved on the training data using 5-fold cross-validation with these parameters is approximately 86.30%.

We can use these optimized parameters to train the KNN model for potentially better performance on the test set.

**Part 2.2 b - Check the percentage of missing values and impute if required**

```python
# Checking the percentage of missing values in each column
missing_percentage = (merged_data.isnull().sum() / len(merged_data)) * 100

# Filtering only columns with missing values
missing_percentage = missing_percentage[missing_percentage > 0]
```

```python
# Imputing the missing values with the mode for the 'LoanOnCard' column (since it's a categorical variable)
if 'LoanOnCard' in missing_percentage.index:
    mode_value = merged_data['LoanOnCard'].mode()[0]
    merged_data['LoanOnCard'].fillna(mode_value, inplace=True)


# Checking again for missing values after imputation
updated_missing_percentage = (merged_data.isnull().sum() / len(merged_data)) * 100
updated_missing_values = updated_missing_percentage[updated_missing_percentage > 0]

missing_percentage, updated_missing_values
```

The initial percentage of missing values for the 'LoanOnCard' column was 0.4%.

After imputation using the mode of the 'LoanOnCard' column, there are no more missing values in the dataset.

**Part 2.2 c - Check for unexpected values in each categorical variable and impute with best suitable value.**

```python
# Identifying the categorical columns
categorical_columns = merged_data.select_dtypes(include=['object']).columns

# Dictionary to store unexpected values for each categorical column
unexpected_values = {}

# Loop through each categorical column to identify unexpected values
for column in categorical_columns:
    # Assuming that valid values for these categorical columns are between 0 and 1 (as they seem to be binary flags)
    unexpected = merged_data[~merged_data[column].isin([0, 1])][column].unique()
    if len(unexpected) > 0:
        unexpected_values[column] = unexpected

unexpected_values
```

```python
# Collecting rows with unexpected values BEFORE imputation
rows_before_imputation = merged_data[merged_data[categorical_columns].isin(unexpected_values).any(axis=1)]
rows_before_imputation


```

```python
# Imputing unexpected values (if any) with the mode for each categorical column
for column, values in unexpected_values.items():
    mode_value = merged_data[column].mode()[0]
    merged_data[column].replace(values, mode_value, inplace=True)
```

```python


# Collecting rows with unexpected values AFTER imputation
rows_after_imputation = merged_data[merged_data[categorical_columns].isin(unexpected_values).any(axis=1)]
rows_after_imputation
```

These values were considered "unexpected" based on the assumption that the valid values for binary flags in the dataset should be 0 or 1.

I've imputed these unexpected values with the mode of their respective columns.

**Part 2.3 a- Split data into X and Y**

```python
# Splitting the merged_data into X (features) and Y (target)
X = merged_data.drop('LoanOnCard', axis=1)
Y = merged_data['LoanOnCard']


print("First 5 rows of X dataframe:\n", X.head())
print("----------------------------------------------------------------------------------------")
print("First 5 rows of Y dataframe:\n", Y.head())
```

**Part 2.3 b - Split data into train and test. Keep 25% data reserved for testing**

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

After running the above code:

X_train and Y_train will contain the features and target variable for the training set, respectively.
X_test and Y_test will contain the features and target variable for the testing set, respectively.

**Part 2.3 c - Train a Supervised Learning Classification base model - Logistic Regression**

```python
# Handle any missing or infinite values in the training and testing sets
missing_columns = X_train.columns[X_train.isnull().any()]
for column in missing_columns:
    median_value = X_train[column].median()
    X_train[column].fillna(median_value, inplace=True)
    X_test[column].fillna(median_value, inplace=True)

# Train the Logistic Regression model

from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_model.fit(X_train, Y_train)

# Getting the training score for evaluation
training_score = logistic_model.score(X_train, Y_train)
training_score
```

The Logistic Regression model has been trained and achieved an accuracy of approximately
91.07
%
on the training data.

All the necessary preprocessing steps were executed to ensure that the data was properly prepared before training the model.

**Part 2.3 d - Print evaluation metrics for the model and clearly share insights.**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Predicting the target values for the test set
Y_pred = logistic_model.predict(X_test)

# Evaluation metrics
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)
confusion = confusion_matrix(Y_test, Y_pred)
report = classification_report(Y_test, Y_pred)


print("Accuracy achieved in the model:",accuracy)
print("Precision achieved in the model:",precision)
print("Recall achieved in the model:",recall)
print("F1 score achieved in the model:",f1)
print("Confusion matrix for the model:\n",confusion)
print("Classification report for the model:\n",report)
```

**Part 2.3 e - Balance the data using the right balancing technique.**

Balancing the data is crucial when dealing with imbalanced datasets, as it can help improve the model's performance, especially on the underrepresented class. There are several techniques to handle imbalanced data:

Oversampling: Increase the number of instances in the underrepresented class by replicating them.

Undersampling: Decrease the number of instances in the overrepresented class by randomly removing them.

SMOTE (Synthetic Minority Over-sampling Technique): Generate new instances in the underrepresented class by interpolating between existing instances.

ADASYN (Adaptive Synthetic Sampling): Similar to SMOTE but focuses on generating samples next to the original samples which are wrongly classified using a k-Nearest Neighbors classifier.

Undersampling can be a straightforward way to balance the data. The idea behind undersampling is to reduce the number of instances from the overrepresented class to make the dataset balanced.

However, it's essential to note that undersampling can lead to a loss of data since we're removing instances from the majority class. In situations where we have a large number of instances in the majority class, this might not be a significant concern, but it's something to be aware of.

```python
from sklearn.utils import resample

# Separate the majority and minority classes
majority_class = merged_data[merged_data['LoanOnCard'] == 0]
minority_class = merged_data[merged_data['LoanOnCard'] == 1]

# Undersample the majority class
majority_undersampled = resample(majority_class,
                                 replace=False,
                                 n_samples=len(minority_class),
                                 random_state=42)

# Combine the undersampled majority class with the minority class to form a balanced dataset
balanced_data = pd.concat([majority_undersampled, minority_class])

# Checking the distribution of the target variable after undersampling
balanced_distribution = balanced_data['LoanOnCard'].value_counts(normalize=True) * 100
balanced_distribution
```

**Part 2.4 b - Tune parameters for each of the models wherever required and finalize a model**

Tuning the parameters of a model can often result in improved performance. Let's proceed with hyperparameter tuning for both SVM and KNN using grid search:

Support Vector Machine (SVM):

C: Regularization parameter. The strength of the regularization is inversely proportional to C. We'll try values like [0.1, 1, 10].

kernel: Specifies the kernel type to be used in the algorithm. We'll try ['linear', 'rbf'].

gamma: Kernel coefficient. We'll try ['scale', 'auto'].

K-Nearest Neighbors (KNN):

n_neighbors: Number of neighbors to use. We'll try values like [3, 5, 7].

weights: Weight function used in prediction. We'll try ['uniform', 'distance'].

algorithm: Algorithm used to compute the nearest neighbors. We'll try ['auto', 'ball_tree', 'kd_tree', 'brute'].

For the sake of computational efficiency, we'll use a subset of potential hyperparameters. Let's start with the grid search for SVM.

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameters for SVM
svm_params = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Grid search for SVM
svm_grid_search = GridSearchCV(SVC(random_state=42), svm_params, cv=5, n_jobs=-1)
svm_grid_search.fit(X_train_balanced, Y_train_balanced)

# Best parameters and score for SVM
svm_best_params = svm_grid_search.best_params_
svm_best_score = svm_grid_search.best_score_

svm_best_params, svm_best_score
```
